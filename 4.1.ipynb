```python
import pandas as pd
import numpy as np
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist,myarr))
panda_list = pd.Series(mylist)
panda_arr = pd.Series(myarr)
panda_dict = pd.Series(mydict)
print(panda_list, "là 1 chuỗi pandas từ list đã cho")
print(panda_arr, "là 1 chuỗi pandas từ array đã cho")
print(panda_dict, "là 1 chuỗi pandas từ dict đã cho")
```
